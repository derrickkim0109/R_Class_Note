```r

# 자치구별 발생 건수의 평균
regionAcc.mean <- aggregate(발생건수 ~ 자치구명, seoul, mean)
head(regionAcc.mean)
```

```r
# regionAcc.mean의 발생건수를 기준으로 내림차순으로 정렬하기
# 1등이 강남구야? 아니? 초딩이니? ---> 표준편차 확인해봐(고딩수준)
arrange(regionAcc.mean, desc(발생건수))
```

```r
# 자치구별 발생건수의 표준편차 구하기 
regionAcc.sd <- aggregate(발생건수 ~ 자치구명, seoul, sd)
head(regionAcc.sd)
```

```r
# regionAcc.sd의 발생건수를 기준으로 내림차순으로 정렬하기
arrange(regionAcc.sd, desc(발생건수))
```

```r
# 표준편차 어디 써먹니?--->>
# 자치구별 발생건수의 변동계수를 구하여라 ( 편차를 구하는 이유는 변동계수(지표값)를 구하기 위해 )
# 변동계수 = 표준편차 / 평균   변동계수 = cv
regionAcc.cv <- regionAcc.sd$발생건수 / regionAcc.mean$발생건수
regionAcc.cv
```

```r
# 자치구명과 변동계수를 가진 DataFrame 만들기
regionAcc <- data.frame(자치구명 = regionAcc.mean$자치구명, 변동계수 = regionAcc.cv)
#head(regionAcc)
# 강남구가 1위이다. 
arrange(regionAcc,변동계수)
```

```r
# 서울시 교통사고 부상자수에 대한 변동계수 구하기
# 1.평균
region.injury.mean <- aggregate(부상자수 ~ 자치구명 ,seoul, mean)
# 2.표준편차
region.injury.sd <- aggregate(부상자수 ~ 자치구명 ,seoul, sd)
# 3.변동계수
region.injury.cv <- region.injury.sd$부상자수 /region.injury.mean$부상자수
# 4.dataframe 만들기
region.injury <- data.frame(자치구명 = region.injury.mean$자치구명,
                            변동계수 = region.injury.cv)
# 5.정렬하기
arrange(region.injury,변동계수)
```

```r
# 교통사고 발생건수와 부상자수의 관계 -> 교통사고가 발생하면 부상자가 꼭 나오나??
# 상관계수 값을 쓰면 된다!
cor(seoul$발생건수, seoul$부상자수)
# 거의 98%상관있다. -> 병원에 간사람. 
```

```r
# 교통사고 발생건수와 사망자수의 관계 -> 
cor(seoul$발생건수, seoul$사망자수)
# 23%---> 관계없다. 

# 상관계수 percentage 
# 0 ~ 0.3 : 상관없음
# 0.3 ~ 0.5 : 약한 상관관계
# 0.5 ~ 0.7 : 상관 있음
# 0.7 ~ : 강한 상관관계가 있음
```

```r
# 교통사고 부상자수와 사망자수의 관계 -> 부상자가 늘어나면 사망자 늘어나??
cor(seoul$부상자수, seoul$사망자수)
#OECD 사망자수는 현장에서 사망한수 
```

```r
# 학렬별 점수 차이
data <- read.csv('../Data/education.csv', header = F)
head(data)
```

```r
str(data)
```

```r
# V1 : 남, 여
# V2 : 점수
# V3 : 국, 영, 수
# V4 : 초, 중, 고, 대, 원
```

```r
data$V1 <- factor(data$V1, levels = 1:2, labels = c('남','여')) # 남과 여니까
data$V3 <- factor(data$V3, levels = 1:3, labels = c('국','영','수')) # 국 영 수
data$V4 <- factor(data$V4, levels = 1:5, labels = c('초','중','고','대','원')) # 초, 중, 고, 대, 원
```

```r
head(data)
```

```r
str(data)
```

```r
# 컬럼 이름 변경
colnames(data) <- c('성별','점수','과목','학력')
head(data)
```

```r
# 막대 그래프로 점수 빈도수 표현
table(data$점수)
```

```r
par(family = "AppleGothic") # for mac


barplot(table(data$점수),
        main ='점수별 빈도수',
        xlab = '점수',
        ylab = '빈도수',
        ylim = c(0,12)
        
       )
```

```r
# 성별, 학력별 빈도수를 표현

table(data$성별)
table(data$학력)
table(data$성별, data$학력)
```

```r
par(family = "AppleGothic") # for mac
barplot(table(data$성별, data$학력),
        main = '성별 학력별 빈도수',
        xlab = '학력별',
        ylab = '빈도수',
        beside = T,
        col = c('orange', 'yellow'),
        ylim = c(0,6),
        legend.text = T, # 레전드는 절대 데이터 가리면 안된다. 
        
       )
```

```r
# 회사 매출현황 데이터 (가상)
#               / 1분기 비교 / 2분기 /   3분기   /  4분기  / 

chart_data <- c(305, 450, 320, 460, 330, 480, 380, 520)
names(chart_data) <- c('2017 1분기','2018 1분기',
                       '2017 2분기','2018 2분기',
                       '2017 3분기','2018 3분기',
                       '2017 4분기','2018 4분기'
                      )
chart_data
```

```r
par(family = "AppleGothic") # for mac
# 매출 현황 비교인데 ylab의 매출이 어느정도인지 구분을 못한다. 
barplot(chart_data,
        main = '2017년도 2018년도 분기별 매출 현황 비교표',
        ylim = c(0,600),

       )
```

```r
par(family = "AppleGothic") # for mac
# 매출 현황 비교인데 ylab의 매출이 어느정도인지 구분을 못한다. 

barplot(chart_data,
        main = '2017년도 2018년도 분기별 매출 현황 비교표',
        ylim = c(0,600),
        col = rainbow(length(chart_data)),
        xlab = '년도별 분기 현황',
        ylab = '매출액(단위:만원)'
       )
```

```r
par(family = "AppleGothic") # for mac
# 옆으로 세워야지 

# ylim -> xlim , xlab -> ylab , ylab -> xlab
barplot(chart_data,
        main = '2017년도 2018년도 분기별 매출 현황 비교표',
        xlim = c(0,600),
        col = rainbow(length(chart_data)),
        ylab = '년도별 분기 현황',
        xlab = '매출액(단위:만원)',
        horiz = T
        
       )
```

```r
# 분기별 간격
par(family = "AppleGothic") # for mac


# ylim -> xlim , xlab -> ylab , ylab -> xlab
barplot(chart_data,
        main = '2017년도 2018년도 분기별 매출 현황 비교표',
        xlim = c(0,600),
        col = rainbow(length(chart_data)),
        ylab = '년도별 분기 현황',
        xlab = '매출액(단위:만원)',
        horiz = T,
        space = 0.5,
        cex.names = 0.6
       )
```

```r
# 분기별, 년도별 
par(family = "AppleGothic") # for mac



barplot(chart_data,
        main = '2017년도 2018년도 분기별 매출 현황 비교표',
        xlim = c(0,600),
        col = rep(c("red","blue"),4),
        ylab = '년도별 분기 현황',
        xlab = '매출액(단위:만원)',
        horiz = T,
        space = rep(c(2, 0.5),4),
        cex.names = 0.6
       )
```
